```python
import torch
from torch import nn
from torch import optim
from torch.utils.data import TensorDataset, DataLoader
from torchmetrics import Accuracy

accuracy = Accuracy(task="multiclass", num_classes=2)
```

```python
class ANN(nn.Module):
    def __init__(self , input_size , hidden_size ,output_size):
        super(ANN, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, hidden_size // 2),
            nn.ReLU(),
            nn.Linear(hidden_size // 2, hidden_size // 4),
            nn.ReLU(),
            nn.Linear(hidden_size // 4, output_size),
        )
    def forward(self, input_tensor):
        res = self.model(input_tensor)
        return res
```

```python
input_size = X_train.shape[1]
hidden_size = 128*2
output_size = 1
model = ANN(input_size, hidden_size, output_size)
```

```python
model.parameters
```

```python
X_train_tensor = torch.from_numpy(X_train) 
y_train_tensor = torch.from_numpy(y_train.values) 
X_test_tensor = torch.from_numpy(X_test)
y_test_tensor = torch.from_numpy(y_test.values)

loss_fn = nn.BCEWithLogitsLoss()


optimizer = optim.Adam(model.parameters() , lr=0.00001)
```

```python
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
epochs = 20
for epoch in range(epochs):
    for X, y in train_loader:
        optimizer.zero_grad()
        X = X.float()
        y_pred = model(X)
        y_reshaped = y.view(-1, 1).float()
        loss = loss_fn(y_pred, y_reshaped)
        loss.backward()
        optimizer.step()
        model.eval()
    with torch.no_grad():
        X_test_tensor = X_test_tensor.float()
        y_pred_logits = model(X_test_tensor)
        y_pred_probs = torch.sigmoid(y_pred_logits)

        y_preds = torch.round(y_pred_probs)

        y_preds = y_preds.squeeze()  
        test_acc = accuracy(y_preds, y_test_tensor)
        print(f"{epoch+1}: Accuracy {test_acc:.4f}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
print(classification_report(y_preds, y_test_tensor))
```

```python
cm = confusion_matrix(y_preds, y_test_tensor)
cf_grp = ConfusionMatrixDisplay(confusion_matrix = cm )
cf_grp.plot()
```

# LSTM , GRU 

```python
from torch.utils.data import TensorDataset, DataLoader
from torchmetrics import Accuracy
accuracy = Accuracy(task="multiclass", num_classes=2)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
X_train_tensor = torch.from_numpy(X_train).to(device)
y_train_tensor = torch.from_numpy(y_train.values).to(device)
X_test_tensor = torch.from_numpy(X_test).to(device)
y_test_tensor = torch.from_numpy(y_test.values).to(device)
```

```python
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
import torch 
import torch.nn as nn
import torch.optim as optim 
```

```python
class LSTM(nn.Module):
    def __init__(self ,input_size , hidden_size, num_class=1 ):
        self.hidden_size = hidden_size
        super(LSTM, self).__init__()
        self.lstm =  nn.LSTM(input_size = input_size , hidden_size = hidden_size*2, num_layers=1 , dropout=0.0, bidirectional=False) 
        self.model = nn.Sequential(
            nn.Linear(hidden_size*2, hidden_size//2),
            nn.ReLU(),
            nn.Linear(hidden_size//2, hidden_size//4),
            nn.ReLU(),
            nn.Linear(hidden_size//4, hidden_size//8),
            nn.ReLU(),
            nn.Linear(hidden_size//8, num_class),
        )
    def forward(self,x):
        x , _ = self.lstm(x)
        x = self.model(x)
        return x 
        
```

```python
input_size = X_train_tensor.shape[1]
hidden_size = 128
model = LSTM(input_size = input_size , hidden_size = hidden_size).to(device)
loss_fn = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters() , lr=0.001)
```

```python
epochs = 20
```

```python
for epoch in range(epochs):
    model.train()
    epoch_loss = 0.0
    for X_train_batch , y_train_batch in train_loader:
        optimizer.zero_grad()
        X_train_batch = X_train_batch.float()
        X_train_batch , y_train_batch = X_train_batch.to(device) , y_train_batch.to(device)
        y_pred = model(X_train_batch)
        y_train_batch_reshaped = y_train_batch.view(-1, 1).float()
        y_train_batch_reshaped = y_train_batch_reshaped.to(device)
        loss = loss_fn(y_pred ,y_train_batch_reshaped)
        loss.backward()
        epoch_loss += loss.item()
        optimizer.step()
    epoch_loss /= len(train_loader)
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {epoch_loss:.4f}")
        
```

```python
model.eval()
test_loss = 0.0
correct_predictions = 0
total_samples = 0
with torch.no_grad():
    for X_test_batch, y_test_batch in test_loader:
        X_test_batch = X_test_batch.float().to(device)
        y_test_batch_reshaped = y_test_batch.view(-1, 1).float().to(device)
        y_pred = model(X_test_batch)
        loss = loss_fn(y_pred, y_test_batch_reshaped)
        test_loss += loss.item() * X_test_batch.size(0)
        predicted_labels = (torch.sigmoid(y_pred) > 0.5).float() 
        correct_predictions += (predicted_labels == y_test_batch_reshaped).sum().item()
        total_samples += X_test_batch.size(0)
test_loss /= total_samples
accuracy = correct_predictions / total_samples
print(f"Test Loss: {test_loss:.4f}, Accuracy: {accuracy:.4f}")
```
